## Cell 1 — Setup

```python
import sys
sys.path.insert(0, "..")


from src.retrieval.fusion import retrieve
from src.retrieval.bm25_retriever import search as bm25_search
from src.retrieval.dense_retriever import search as dense_search

print("Imports OK")
```

## Cell 2 — BM25 Smoke Test

```python
query = "INA trial Red Fort defendants"

results = bm25_search(query, top_k=5)
print(f"BM25 results for: {query!r}")
for r in results:
    print(f"[{r['bm25_score']:.3f}] {r['doc_id']} | {r['bias_tag']} | {r['text'][:150]}")
```

## Cell 3 — Dense Smoke Test

```python
results = dense_search(query, top_k=5)
print(f"Dense results for: {query!r}")
for r in results:
    print(f"[{r['dense_score']:.4f}] {r['doc_id']} | {r['language']} | {r['text'][:150]}")
```

## Cell 4 — Fused Retrieval (RRF)

```python
results = retrieve(query, top_k=10)
print(f"Fused RRF results for: {query!r}")
for r in results:
    source = "BOTH" if r["in_bm25"] and r["in_dense"] else ("BM25" if r["in_bm25"] else "Dense")
    print(f"[{r['rrf_score']:.4f}] {source:6s} | {r['bias_tag']:10s} | {r['language']:10s} {r['text'][:120]}")
```

## Cell 5 — Bias Filter Test

```python
results = retrieve(query, top_k=5, filters={"bias_tags": ["british_legal"]})
print("british_legal only:")
for r in results:
    print(f"  {r['bias_tag']} | {r['text'][:100]}")

results = retrieve(query, top_k=5, filters={"bias_tags": ["ina_testimony"]})
print("ina_testimony only:")
for r in results:
    print(f"  {r['bias_tag']} | {r['text'][:100]}")
```

## Cell 6 — Language Filter Test

```python
results = retrieve("صوبہ دار شاہ نواز خان", top_k=5, filters={"languages": ["te"]})
print("Telegu query, Urdu chunks only:")
for r in results:
    print(f"  [{r['language']}] {r['text'][:150]}")

if not results:
    print("  (no Urdu chunks found — check language_detected values in your data)")
```
